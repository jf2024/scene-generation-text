```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import google.generativeai as genai
from google.generativeai.types import HarmCategory, HarmBlockThreshold

# 1. Load your data
df = data.copy()

# 2. Create embeddings for all scripts
model = SentenceTransformer('all-MiniLM-L6-v2')  # Or another embedding model
df['script_embedding'] = df['raw_script'].apply(lambda x: model.encode(x))  # Truncate if needed

# 3. Configure Gemini API
genai.configure(api_key='......')

# 4. Function to find similar scripts based on creator
def get_similar_scripts(writer=None, director=None, genre=None, top_k=3):
    # Filter by metadata if provided
    filtered_df = df.copy()
    if writer:
        filtered_df = filtered_df[filtered_df['writers'].str.contains(writer, na=False)]
    if director:
        filtered_df = filtered_df[filtered_df['directors'].str.contains(director, na=False)]
    if genre:
        filtered_df = filtered_df[filtered_df['genres'].str.contains(genre, na=False)]
    
    # If no matches, return to the full dataset but note this
    if len(filtered_df) == 0:
        filtered_df = df.copy()
        print(f"No exact matches found, returning similar content from all scripts")
    
    # Return the top_k samples
    return filtered_df.head(top_k)[['title', 'writers', 'directors', 'raw_script']]

# 5. Function to generate a scene without character input
def generate_scene(writer=None, director=None, genre=None, setting=None):
    # Get relevant script examples
    similar_scripts = get_similar_scripts(writer, director, genre)
    
    # Create examples context (truncated for prompt size)
    examples = ""
    for _, row in similar_scripts.iterrows():
        examples += f"Example from {row['title']} by {row['writers']}, directed by {row['directors']}:\n"
        examples += row['raw_script'][:500] + "...\n\n"  # First 500 chars as example
    
    # Build the prompt
    prompt = f"""
    Based on the following examples of writing and directing style, create a new original scene.
    
    {examples}
    
    Please write a scene with:
    Writer style: {writer if writer else 'Any appropriate style'}
    Director style: {director if director else 'Any appropriate style'}
    Genre: {genre if genre else 'Any appropriate genre'}
    Setting: {setting if setting else 'Any appropriate setting'}
    
    Format the scene as a proper screenplay with scene heading, action paragraphs, character names, and dialogue.
    """
    
    # Configure Gemini model settings
    generation_config = {
        "temperature": 0.9,
        "top_p": 1,
        "top_k": 32,
        "max_output_tokens": 2048,
    }
    
    # safety_settings = {
    #     HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    #     HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    #     HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    #     HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    # }
    
    # Generate using Gemini
    model = genai.GenerativeModel(
        model_name="gemini-2.0-flash",
        generation_config=generation_config,
        #safety_settings=safety_settings
    )
    
    response = model.generate_content(prompt)
    
    return response.text

# Example usage
new_scene = generate_scene(
    writer="Quentin Tarantino",
    director="Christopher Nolan",
    setting="A rainy night at an abandoned school with a young 21 year old adult make having a knife, thinking about his existence and purpose in a depressive state",
    genre="Thriller"
)
```
