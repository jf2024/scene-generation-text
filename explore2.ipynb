```python
# Directory with raw scripts
script_folder = "data/raw_texts/raw_texts"  # Update path accordingly
script_dict = {}

# Go through each script file
for filename in os.listdir(script_folder):
    if filename.endswith(".txt"):
        imdb_id = filename.split("_")[-1].replace(".txt", "")
        file_path = os.path.join(script_folder, filename)

        try:
            with open(file_path, "r", encoding="utf-8") as f:
                script_dict[imdb_id] = f.read()
        except UnicodeDecodeError:
            # Try again with a fallback encoding
            with open(file_path, "r", encoding="latin-1") as f:
                script_dict[imdb_id] = f.read()

# Match using formatted IMDb IDs (padded with zeros)
data["imdbid_str"] = data["imdbid"].astype(str).str.zfill(7)
data["raw_script"] = data["imdbid_str"].map(script_dict)

# Clean up
data.drop(columns=["imdbid_str"], inplace=True)
```
